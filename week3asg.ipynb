```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Palmer Penguins dataset from seaborn's built-in datasets
penguins = sns.load_dataset('penguins')

# Display the first few rows of the dataset
print(penguins.head())

# Scatter plot of bill length vs bill depth
plt.figure(figsize=(10, 6))
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Bill Length vs Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()

# Histogram of flipper length
plt.figure(figsize=(10, 6))
sns.histplot(penguins['flipper_length_mm'].dropna(), bins=30, kde=True)
plt.title('Flipper Length Distribution')
plt.xlabel('Flipper Length (mm)')
plt.show()

# Bar plot of species counts by island
plt.figure(figsize=(10, 6))
sns.countplot(data=penguins, x='island', hue='species')
plt.title('Species Counts by Island')
plt.xlabel('Island')
plt.ylabel('Count')
plt.show()
```
