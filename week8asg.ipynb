```python
train.info()
```

```python
x=train.iloc[:,:-1]
x.drop(columns=['Loan_ID'],inplace=True)
y=train.loc[:,'Loan_Status']
y=y.map({'Y':1,'N':0})
```

```python
obj_col = x.select_dtypes(include=['object']).columns
int_col = x.select_dtypes(include=['int64','float64']).columns
```

```python
from sklearn.impute import SimpleImputer

imp1 = SimpleImputer(strategy='mean')
imp1.fit(x[int_col])
x[int_col] = imp1.transform(x[int_col])

imp2 = SimpleImputer(strategy='most_frequent')
imp2.fit(x[obj_col])
x[obj_col] = imp2.transform(x[obj_col])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in obj_col:
    x[col] = le.fit_transform(x[col])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb
from catboost import CatBoostClassifier

# Define the models
xgb_model = xgb.XGBClassifier()
catboost_model = CatBoostClassifier()

# Define the parameter grids
xgb_param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

catboost_param_grid = {
    'iterations': [500],
    'learning_rate': [0.01, 0.1],
    'depth': [4, 6],
    'l2_leaf_reg': [1, 3],
    'border_count': [32, 50]
}

# Perform randomized search
xgb_random_search = RandomizedSearchCV(xgb_model, param_distributions=xgb_param_grid, n_iter=20, cv=3, verbose=1, n_jobs=-1)
catboost_random_search = RandomizedSearchCV(catboost_model, param_distributions=catboost_param_grid, n_iter=20, cv=3, verbose=1, n_jobs=-1)

# Fit the models
xgb_random_search.fit(x, y)
catboost_random_search.fit(x, y)
```

```python
# Perform the fitting
xgb_random_search.fit(x, y)
catboost_random_search.fit(x, y)

# Check if fitting was successful
if hasattr(xgb_random_search, 'best_estimator_'):
    xgb_best = xgb_random_search.best_estimator_
else:
    print("XGBoost Random Search did not complete successfully.")

if hasattr(catboost_random_search, 'best_estimator_'):
    catboost_best = catboost_random_search.best_estimator_
else:
    print("CatBoost Random Search did not complete successfully.")
```

```python
from sklearn.metrics import log_loss

# Predict probabilities and calculate log loss for XGBoost
y_pred_xgb = xgb_best.predict_proba(x)[:, 1]
loss_xgb = log_loss(y, y_pred_xgb)
print(f'Binary Cross-Entropy Loss for XGBoost: {loss_xgb}')

# Predict probabilities and calculate log loss for CatBoost
y_pred_catboost = catboost_best.predict_proba(x)[:, 1]
loss_catboost = log_loss(y, y_pred_catboost)
print(f'Binary Cross-Entropy Loss for CatBoost: {loss_catboost}')
```
